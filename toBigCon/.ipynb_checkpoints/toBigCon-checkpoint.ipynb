## 문제 출제의 목적
1. 봄, 여름, 가을, 겨울 시즌별 운영상품 및 카테고리는 반복되는 상황이지만, 외부요인은 날씨, 사회이슈 시청률에 따라 상품 매출 실적의 변환가 있었다.
2. 내부요인과 외부요인을 통해 프로그램 매출 실적을 사전에 예측하고, 대응하자

## 주제 
1. 데이터를 활용하여 20년 6월 1일~30일 프로그램 상품 판매실적을 예측하고 최적수익을 고려한 요일별 / 시간대별 / 카테고리별 편성 최적화 방안 제시

## 용어 설명
1. 판매가 0원 은 추정 제외
2. 매주 토요일은 정보방송시간으로 추정 제외
3. 편성 듀레이션은 10~60분 내외
4. 편성표는 6:20~2:20 으로 구성 (심야시간 제외)
5. 취급액 = 판매단거 X 주문량

6. 예측상품중 과거 실적이 없는 경우는 유사 카테고리 혹은 머더코드로 예측

## 제공데이터
1. 실적데이터 : 매출 실적데이터
2. 외부데이터 : 요일별/시간대별 분단위 시청률
3. 예측시점 : 20년6월1일~30일 프로그램별 분당 실적예측

### 고민된점
1. 노출(분) 이 Nan 값인건 뭐징
- 동일한 머더코드, 방송일시, 상품군, 판매단가면 같은거로 채워도 될거같다. 
- 그 앞의 값으로 채워넣으면 같은 효과

```python
df['노출(분)'] = df['노출(분)'].fillna(method='ffill')
```
