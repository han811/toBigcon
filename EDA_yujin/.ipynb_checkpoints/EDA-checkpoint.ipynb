## 결측치 확인

```python
performance_data.isna().sum()
```

노출(분)과 취급액의 결측치가 좀 됨

### 취급액 결측치

```python
cond = performance_data['취급액'] == 50000
cond2 = performance_data['판매단가'] > 50000
```

```python
performance_data[cond & cond2] # 얘넨 취급액을 다 0으로 바꿔줘야 함
```

```python
performance_data[performance_data['취급액'].isna()] # 판매단가가 0인 경우
```

### 판매단가 결측치

```python
performance_data.loc[performance_data['판매단가'] <= 0, "상품군"].unique()
```

```python
performance_data.loc[performance_data['상품군'] == "무형", "판매단가"].sum()
```

판매단가가 0일 때 상품군은 모두 무형임. 그 역도 성립

### 노출 결측치

```python
exposure_na_idx = performance_data[performance_data['노출(분)'].isna()].index
```

```python
for count, idx in enumerate(exposure_na_idx):
  if count > 100:
    break
  print(performance_data.iloc[range(idx-2, idx+3), :])
```

완벽하게 확인을 한 것은 아니나, 노출 시간의 결측치는 같은 시간대의 다른 상품코드일 확률이 높다. (즉 그냥 방송 같이 묶어서 하는 다른 상품)
